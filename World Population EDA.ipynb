```python
for x in data["Continent"].unique():
    temp=data[data["Continent"]=="Asia"][["Country/Territory","2022 Population"]].sort_values(by="2022 Population",ascending=False,ignore_index=False)[:7]
    plt.figure(figsize=(8,4))
    plt.title(f"Top 5 Countries based on 2022 Population in {x}")
    sns.barplot(data=temp,x="Country/Territory",y="2022 Population")
    plt.show()
```

```python
diff_70_22=data.groupby("Continent")[["1970 Population","2022 Population"]].sum().sort_values(by="Continent",ascending=True).reset_index()
```

```python
diff_70_22
```

```python
plt.figure(figsize=(16,8))
plt.plot(diff_70_22["Continent"],diff_70_22["1970 Population"],label="1970",color="red",marker="o")
plt.plot(diff_70_22["Continent"],diff_70_22["2022 Population"],label="2022",color="blue",marker="x")
plt.xlabel("Continent")
plt.ylabel("Population")
plt.grid()
plt.legend()
plt.show()
```
